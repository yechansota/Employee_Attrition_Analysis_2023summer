### ¬∑ Prepocessing the Data Modeling
---

````python
# ‚ñ∂ Define numerical and categorical value from the datamart
numeric_list=[]
categoical_list=[]

for i in df.columns :
  if df[i].dtypes == 'O' :
    categoical_list.append(i)
  else :
    numeric_list.append(i)

print("categoical_list :", categoical_list)
print("numeric_list :", numeric_list)
````

**Categoical_list** : ['BusinessTravel', 'Department',
'EducationField', 'Gender', 'JobRole', 'MaritalStatus', 'Over18', 'OverTime', '*Age_gp*', '*Role_Company_gp*']

**Numeric_list** : ['Age', '*Attrition*', 'DailyRate', 'DistanceFromHome', 'Education', 'EmployeeCount', 'EmployeeNumber', 'EnvironmentSatisfaction', 'HourlyRate', 'JobInvolvement', 'JobLevel', 'JobSatisfaction', 'MonthlyIncome', 'MonthlyRate', 'NumCompaniesWorked', 'PercentSalaryHike', 'PerformanceRating', 'RelationshipSatisfaction', 'StandardHours', 'StockOptionLevel', 'TotalWorkingYears', 'TrainingTimesLastYear', 'WorkLifeBalance', 'YearsAtCompany', 'YearsInCurrentRole', 'YearsSinceLastPromotion', 'YearsWithCurrManager', 'Role_Company']

````python
#imports machine learning alogrithm used for classfication task
#import various ML functions
# Before starting to model, it's important to distinguish between variables X and Y.
# This will help ensure accuracy and precision when analyzing and interpreting the data.
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics

X=df.drop(['Attrition', 'Age_gp', 'Role_Company_gp', 'Role_Company'], axis=1)
Y=df['Attrition']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, stratify=Y)

print(x_train.shape)
print(y_train.shape)

print(x_test.shape)
print(y_test.shape)
````

````python
from sklearn.preprocessing import LabelEncoder #label encoding for categorical variables = le

for col in categoical_list:
    print(col)
    le = LabelEncoder() #col = categorical column
    le.fit(list(x_train[col].values) + list(x_test[col].values))
    x_train[col] = le.transform(x_train[col]) #Replaces the categorical values with numerical labels.
    x_test[col] = le.transform(x_test[col])
````

````python
x_test[categoical_list].head(5)
````

### ¬∑Model Evaluation
---

````python
from sklearn.metrics import classification_report
rfc = RandomForestClassifier(random_state = 123456, max_depth=3, n_estimators=125)
# The random forest model will produce the same results whenever the code is executed with the same random_state value.
# By limiting the depth of the trees, max_depth helps the model generalize better and produce more accurate predictions on unseen data.
# This will be composed of 125 decision trees.
rfc.fit(x_train, y_train)

# ÏòàÏ∏°

y_pred_train = rfc.predict(x_train)
y_pred_test = rfc.predict(x_test)

print(classification_report(y_train, y_pred_train))
print(classification_report(y_test, y_pred_test))
````

````python
y_pred_train
````

````python
pd.Series(y_pred_test).value_counts()
# The model has classified the majority of the test samples as class 0 (negative class)
# only a very small number of samples as class 1 (positive class) based on the chosen threshold.
# --> This suggests that the model is more confident in assigning these samples to the negative class (class 0) based on their corresponding predicted probabilities.
````

````python
from sklearn.metrics import classification_report

y_pred_train = rfc.predict_proba(x_train)[:, 1]
# If the probability is greater than or equal to 0.3,
# It sets the corresponding value in y_pred_train to 1, representing a positive prediction.
for i in range(0,len(y_pred_train)):
    if y_pred_train[i]>=.3:
       y_pred_train[i]=1
    else:
       y_pred_train[i]=0

y_pred_test = rfc.predict_proba(x_test)[:, 1]
for i in range(0,len(y_pred_test)):
    if y_pred_test[i]>=.3:
       y_pred_test[i]=1
    else:
       y_pred_test[i]=0

print(classification_report(y_train, y_pred_train))
print(classification_report(y_test, y_pred_test))
````

````python
# Optimizing relevant variables prevents overfitting in both training and new, unseen data.
# The result of train/test indicate the model perform well in general
from sklearn.metrics import roc_auc_score

y_pred_train_proba = rfc.predict_proba(x_train)[:, 1]
y_pred_test_proba = rfc.predict_proba(x_test)[:, 1]


roc_score_train = roc_auc_score(y_train, y_pred_train_proba)
roc_score_test = roc_auc_score(y_test, y_pred_test_proba)

print("roc_score_train :", roc_score_train)
print("roc_score_test :", roc_score_test)
````

````python
from sklearn.metrics import roc_curve
def roc_curve_plot(y_test , pred_proba_c1):
    # Transfer FPR and TPR based on the number of threshold
    #Draw the ROC cuve and dignoal line on the graph
    fprs , tprs , thresholds = roc_curve(y_test ,pred_proba_c1)

    plt.plot(fprs , tprs, label='ROC')
    plt.plot([0, 1], [0, 1], 'k--', label='Random', color='red')

    # Change the X-axis scale of the FPR to 0.1 units and set the X and Y-axis names. Why? -->
    start, end = plt.xlim()
    plt.xticks(np.round(np.arange(start, end, 0.1),2))
    plt.xlim(0,1)
    plt.ylim(0,1)
    plt.xlabel('FPR( 1 - Sensitivity )')
    plt.ylabel('TPR( Recall )')
    plt.legend()
    plt.show()
````
